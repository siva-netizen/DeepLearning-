```python
import pandas as pd
import seaborn as sns
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
cls_0 = 'human detection dataset/0'
cls_1 = 'human detection dataset/1'
```

```python
import os
import cv2
from os import listdir
```

```python
set_a = set()
for filename in listdir(cls_0):
    img = cv2.imread(os.path.join(cls_0, filename))
    if img.shape not in set_a:
        set_a.add(img.shape)
for a in set_a:
    print(a)
```

```python
set_b = set()
for filename in listdir(cls_1):
    img = cv2.imread(os.path.join(cls_1, filename))
    if img.shape not in set_b:
        set_b.add(img.shape)
for a in set_b:
    print(a)
```

```python

parent_folder = 'human detection dataset'
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator


datagen = ImageDataGenerator(rescale=1./150)
data = datagen.flow_from_directory(
    parent_folder,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary',      
    shuffle = False
)
print(data.class_indices)  
```

```python
i=80
while i < 95:
    for images, labels in data:
        print(f"Image shape: {images[0].shape}")  # Shape of the first image
        print(f"Label: {labels[0]}")             # Label of the image

        # Display the image
        plt.imshow((images[0] * 150).astype('uint8'))  # Rescale to 0-255 range for display
   # Convert to uint8 for visualization
        plt.axis('off')
        plt.show()

        break  # Inspect only the first image
    i+=1
```

```python
from sklearn.model_selection import train_test_split
all_images = []
all_labels = []
for images, labels in data:
    all_images.append(images)
    all_labels.append(labels)
    if len(all_images) * data.batch_size >= data.samples:
        break
```

```python
# Convert lists to numpy arrays
all_images = np.concatenate(all_images, axis=0)
all_labels = np.concatenate(all_labels, axis=0)
```

```python
x_train, x_test, y_train, y_test = train_test_split(all_images,all_labels, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
Sequential_model = Sequential(
    [
        Conv2D(32,(3,3),strides=1,activation='relu',input_shape=(150,150,3)),
        Conv2D(64,(3,3),strides=1,activation='relu'),
        MaxPooling2D(pool_size=(2,2),strides=2),
        Conv2D(128,(3,3),strides=1,activation='relu'),
        Flatten(),
        Dense(128,activation='relu'),
        Dense(64,activation='relu'),
        Dense(32,activation='relu'),
        Dense(1,activation='sigmoid')
    ]
    )
```

```python
Sequential_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history  = Sequential_model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
Sequential_model.summary()
```

```python
Sequential_model.save('model.h5')
```

```python
Sequential_model.evaluate(x_test,y_test)
```
