```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
fashion_minist = tf.keras.datasets.fashion_mnist
(train_image,train_labels),(test_images,test_labels) = fashion_minist.load_data()
```

Normalize the data
1. Pixel Value Range in Images
Pixel values in grayscale images typically range from 0 to 255.
Dividing by 255.0 scales these values to the range [0, 1].
2. ***Prepares for Activation Functions***
Many activation functions like sigmoid, tanh, or ReLU are sensitive to input value ranges.
Input values scaled to [0, 1] help these functions operate effectively and avoid saturation issues (where gradients become very small).

```python
train_image = train_image / 255
test_images = test_images / 255
```

```python
train_image.min(),train_image.max()
```

```python
# Transform the output labels to categorical format before splitting
from tensorflow.keras.utils import to_categorical
train_labels = to_categorical(train_labels, num_classes=10)
test_labels = to_categorical(test_labels, num_classes=10)
```

```python
from sklearn.model_selection import train_test_split
train , val , train_label , val_label = train_test_split(train_image,train_labels,random_state=42,train_size=0.8)
```

```python
from tensorflow import keras
```

```python
train.shape,test_images.shape,val.shape
```

```python
x_train = train.reshape(48000, 784)
x_valid = val.reshape(12000, 784)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Input,Dropout
from tensorflow.keras.regularizers import l1_l2
```

Step 3: Fine-tuning the Regularization Strength
Higher

位: A larger value of

位 (e.g., 0.1 or 1.0) will impose a stronger penalty on the weights, which could lead to underfitting if too large.
Lower

位: A smaller value of

位 (e.g., 0.001) will apply a weaker penalty, which could lead to overfitting if too small.

```python
model = Sequential(
    [
        Input(shape=(784,)),
        Flatten(),
        Dense(256,activation='relu',kernel_regularizer=l1_l2(l1=0.01,l2=0.05)),
        Dropout(0.1),
        Dense(128,activation='relu',kernel_regularizer=l1_l2(l1=0.01,l2=0.05)),
        Dropout(0.1),
        Dense(64,activation='relu',kernel_regularizer=l1_l2(l1=0.01,l2=0.05)),
        Dropout(0.1),
        Dense(10,activation='softmax')
    ]
)
```

image.png

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python


# Compiling the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
early_stopping = EarlyStopping(
    monitor = 'val_loss',
    patience = 5,
    restore_best_weights = True
)

# Fit the model
history = model.fit(
    x_train, train_label,
    epochs=50,
    verbose=1,
    validation_data=(x_valid, val_label),
    callbacks = [early_stopping]
)
```

```python
model.save('fasionminst.keras')
```

```python
# Reshape the test images to (10000, 784)
test_images = test_images.reshape(10000, 784)
# Evaluate the model on the test data
model.evaluate(test_images, test_labels,verbose=2)
```

***How to Fix This***
1. **Reduce L1 Penalty**:
Use a smaller L1 penalty (e.g., l1=0.01) to avoid excessive sparsity.
L2 regularization can be kept at 0.05, as it stabilizes the optimization.
2. **Lower Dropout Rate**:
Reduce dropout to 0.1 or remove it entirely when using L1 regularization.
3. **Increase Patience in EarlyStopping**:
Set patience to a higher value (e.g., 5-10) to give the model more time to converge.

1. Possible Issues Leading to 100% Accuracy
Potential Issue	Explanation
Data Leakage	If some part of your test data is inadvertently included in the training or validation sets, the model "sees" the data during training and achieves artificially high accuracy.
Simple Dataset	If your dataset is very small or has repetitive patterns, the model might memorize it instead of generalizing.
Too High Regularization/Dropout	High penalties or dropout can drive the model to memorize the limited data effectively, as it ignores underlying patterns.
Overfitting to Noise	With small datasets or excessive training, the model may overfit random noise instead of true patterns, appearing perfect on test data.
Data Preprocessing Issue	Misalignment in preprocessing steps (e.g., improper normalization or augmentation applied inconsistently) can make the test set unintentionally "easy" for the model.
